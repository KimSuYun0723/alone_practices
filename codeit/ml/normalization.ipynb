### 1st practice

```python
NBA_FILE_PATH = "C:/_SY/alone_practice/codeit/ml/NBA_player_of_the_week.csv"
nba_player_of_the_week_df = pd.read_csv(NBA_FILE_PATH)
```

```python
nba_player_of_the_week_df.describe()
```

```python
height_weight_age_df = nba_player_of_the_week_df[['Height CM', 'Weight KG', 'Age']]
height_weight_age_df.head()
```

```python
scaler1 = preprocessing.MinMaxScaler() # Min-Max Normalization : min=0, max=1
normalized_data = scaler1.fit_transform(height_weight_age_df)

scaler2 = preprocessing.StandardScaler() # Standardization : m=0, sigma=1
standarized_data = scaler2.fit_transform(height_weight_age_df)

print(normalized_data)
standarized_data
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=['Height', 'Weight', 'Age'])
standarized_df = pd.DataFrame(standarized_data, columns=['Height', 'Weight', 'Age'])
```
