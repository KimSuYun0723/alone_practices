```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures

from math import sqrt

import numpy as np 
import pandas as pd
```

```python
ADMISSION_FILE_PATH = 'C:/_SY/alone_practice/codeit/ml/admission_data.csv'
admission_df = pd.read_csv(ADMISSION_FILE_PATH).drop('Serial No.', axis=1)
```

```python
admission_df.head()
```

```python
X = admission_df.drop(['Chance of Admit '], axis=1)

polynominal_transformer = PolynomialFeatures(6)
polynominal_features = polynominal_transformer.fit_transform(X.values)
features = polynominal_transformer.get_feature_names_out((X.columns))

X = pd.DataFrame(polynominal_features, columns=features)
```

```python
X.head()
```

```python
y = admission_df[['Chance of Admit ']]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
y_test_predict = model.predict(X_test)
```

```python
mse = mean_squared_error(y_train, y_train_predict)

print("training set에서의 성능")
print("------------------------")
print(sqrt(mse))

mse = mean_squared_error(y_test, y_test_predict)

print("test set에서의 성능")
print("------------------------")
print(sqrt(mse))

# overfitting 되었다는 뜻
```
