```python
#####################################################

speech_data = []
idx =1

for page_num in range(2,11):
    driver.get('https://www.korea.kr/briefing/speechList.do')
    time.sleep(3)
    
    if page_num != 1:
        page_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f'//*[@id="container"]/div/div/div/div[4]/span[{page_num}]/a')))

        driver.execute_script("arguments[0].scrollIntoView();", page_button)
        time.sleep(1)
        driver.execute_script("arguments[0].click();", page_button)  # JavaScript로 클릭

    # 30 titles crawling
    for title_num in range(1,31):
        o_title = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[2]/a')))
        o_date = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[4]')
        o_speaker = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[3]')
        title = o_title.text
        date = o_date.text
        speaker = o_speaker.text
        print(title)

        # Crawling each titles
        try:
            driver.execute_script("arguments[0].click();", o_title)  # JavaScript로 클릭
            time.sleep(1)
        except ElementClickInterceptedException:
            print(f"Click intercepted at title {title_num}, retrying with JavaScript click")
            driver.execute_script("arguments[0].click();", o_title)  # JavaScript로 재시도

        o_title.click()
        o_speech = driver.find_element(By.XPATH, '//*[@id="container"]/div/div/div[2]/div' )
        speech = o_speech.text

        d = {"index":idx, "title":title, "date":date, "speaker":speaker, "speech": speech}
        speech_data.append(d)
        idx+=1

        # Go back to the page of 30 titles
        driver.back()
        time.sleep(3)
```

```python
speech_data = []
idx = 1

for page_num in range(1, 11):
    driver.get('https://www.korea.kr/briefing/speechList.do')
    time.sleep(3)
    
    if page_num != 1:
        page_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f'//*[@id="container"]/div/div/div/div[4]/span[{page_num}]/a')))
        
        driver.execute_script("arguments[0].scrollIntoView();", page_button)
        time.sleep(1)
        driver.execute_script("arguments[0].click();", page_button)  # JavaScript로 클릭

    # 30 titles crawling
    for title_num in range(1, 31):
        attempts = 0
        while attempts < 3:
            try:
                o_title = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[2]/a')))
                o_date = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[4]')
                o_speaker = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[3]')
                title = o_title.text
                date = o_date.text
                speaker = o_speaker.text
                print(title)
                
                driver.execute_script("arguments[0].scrollIntoView();", o_title)
                time.sleep(1)

                # Crawling each titles
                driver.execute_script("arguments[0].click();", o_title)  # JavaScript로 클릭
                o_speech = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div/div/div[2]/div')))
                speech = o_speech.text

                d = {"index": idx, "title": title, "date": date, "speaker": speaker, "speech": speech}
                speech_data.append(d)
                idx += 1

                # Go back to the page of 30 titles
                driver.back()
                time.sleep(3)
                break
            
            except StaleElementReferenceException:
                print(f"StaleElementReferenceException occurred for title {title_num}, retrying.")
                attempts += 1
                time.sleep(1)
                driver.refresh()  # 페이지 새로고침 후 재시도
                continue

driver.quit()
```

```python
driver = webdriver.Chrome() 

speech_data = []

def click_right(driver):
    next_button = driver.find_element(By.XPATH, '//*[@id="container"]/div/div/div/div[4]/a[3]')
    next_button.click()
    time.sleep(3)

right_num = 1


try:
    while True:
        # 드라이버 초기화 및 페이지 이동
        driver.get('https://www.korea.kr/briefing/speechList.do')

        # ">" 클릭 (첫 번째 반복이 아닌 경우)
        if right_num != 1:
            for _ in range(right_num-1):
                click_right(driver)
                time.sleep(3)
        
        try:
            # 페이지 반복
            for page_num in range(1, 11):  # 섹션당 10페이지가 있다고 가정
                # 페이지 번호 클릭
                page_button = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/div[4]/span[{page_num}]/a')
                page_button.click()
                time.sleep(3)

                for title_num in range(1, 31):  # 페이지당 30개의 제목이 있다고 가정
                    try:
                        # 페이지의 제목, 날짜, 연설자 크롤링
                        o_title = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[2]/a')
                        o_date = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[4]')
                        o_speaker = driver.find_element(By.XPATH, f'//*[@id="container"]/div/div/div/table/tbody/tr[{title_num}]/td[3]')
                        title = o_title.text
                        date = o_date.text
                        speaker = o_speaker.text

                        # 제목 클릭
                        o_title.click()
                        time.sleep(3)

                        # 해당 제목의 연설 텍스트 크롤링
                        o_speech = driver.find_element(By.XPATH, '//*[@id="container"]/div/div/div[2]/div')
                        speech = o_speech.text

                        d = {"title": title, "date": date, "speaker": speaker, "speech": speech}
                        speech_data.append(d)
                        print(1)
                        print(speech_data)
                        driver.back()
                        time.sleep(3)

                    except (ElementClickInterceptedException, NoSuchElementException) as e:
                        print(f"페이지 {page_num}의 제목 {title_num}에서 예외 발생: {str(e)}")
                        continue
                right_num+=1
        except (NoSuchElementException):
            print("NoSuchElementException - page 부족")
except NoSuchElementException:
    print("NoSuchElementException - 다음 right_num 부족")
finally:
    driver.quit()
```
